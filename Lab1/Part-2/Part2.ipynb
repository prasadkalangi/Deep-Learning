### a) Gradient Descent

Gradient descent is an optimization algorithm which is commonly-used to train machine learning models and neural networks. These models gain knowledge over time by using training data, and the cost function in gradient descent especially serves as a barometer by assessing the accuracy of each iteration of parameter changes. The model will keep changing its parameters to provide the minimal error until the function is close to or equal to zero.

there are different type of gradient descent:

1. **Batch Gradient Descent:**
    It updates the parameters using the average gradient of the cost function for the entire training set. It calculates the gradient of the cost function with respect to parameters for all training examples.It guarantees the convergence to the global minimum, though, it is computationally expensive and slow for large datasets.

2. **Stochastic Gradient Descent (SGD):**
    It updates the  parameters using the gradient of one training example at a time, randomly selects a training example, computes its gradient, and updates parameters.It is computationally efficient, can converge faster than batch gradient descent.

3. **Mini-Batch Gradient Descent:**
    It updates the parameters using the gradient of a small subset (mini-batch) of the training set, calculates the average gradient of the cost function for the mini-batch and updates parameters.

### b) Drop Out

 It is a regularization method used to avoid overfitting problems during training. In this strategy, during each forward and backward run in the training phase, a random subset of neurons and their connections are dropped out or momentarily disregarded. This stochastic approach helps the neural network learn more robust features that generalize well to new input by preventing it from being overly dependent on certain neurons. Dropout, which aids in improved generalization, essentially replicates training an ensemble of various networks using shared parameters. Dropout is normally disabled during inference, and predictions are made by using the entire network. It is a popular regularization strategy in many deep learning architectures due to its simplicity, computational effectiveness, and effectiveness in reducing overfitting.

### c) Activation Function

An activation function in neural networks is a mathematical function that is applied to every neuron's output. Depending on the input it gets, it decides whether or not to activate a neuron.

The neural network may learn more intricate and nuanced interactions between the input and output because to the nonlinearity that activation functions bring. A neural network would perform similarly to a straightforward linear regression model in the absence of an activation function.

Neural networks utilize a variety of activation function types, like as

1) SIGMOID : <br>
One mathematical function that is frequently utilized in neural networks is the sigmoid activation function. With the help of this nonlinear activation function, a real-valued input may be changed to a value between 0 and 1.

### e) Epochs, Iterations and Batch size

Period During the training process, an epoch is the full run of the dataset.Every training example in the dataset is used to train the model during an epoch, and the weights of the model are modified according to the discrepancy between each sample's true and anticipated outputs.

A single training cycle, in which a batch of input data is fed into the model, is referred to as an iteration.

Size of Batch The amount of training instances or samples that a neural network processes in a single forward/backward pass is referred to as the batch size.


The input data is often split into batches and sent through the neural network in a sequential manner during training. The weights and biases of the network are updated depending on the average loss calculated across the whole batch, which is composed of a specific amount of input samples every batch.

The training procedure may be significantly impacted by the batch size. A lower batch size may lead to more frequent changes to the network's weights and biases, accelerating convergence and preventing overfitting. Smaller batch sizes, however, may also lengthen the training period and add to gradient noise, which would cause the convergence to happen more slowly.

However, a higher batch size may yield a more stable gradient estimate, resulting in quicker training periods and smoother convergence. Larger batches, however, may need more memory and processing power and may lead to fewer updates to the network's weights and biases, which might hinder convergence and increase the danger of overfitting.

The ideal batch size is determined by a number of variables, including the job at hand, the amount of the dataset, the computational resources that are available, and the complexity of the model.

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt

def f1(y_true, y_pred):
    def recall(y_true, y_pred):
        true_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_true * y_pred, 0, 1)))
        possible_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_true, 0, 1)))
        recall = true_positives / (possible_positives + keras.backend.epsilon())
        return recall

    def precision(y_true, y_pred):
        true_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_true * y_pred, 0, 1)))
        predicted_positives = keras.backend.sum(keras.backend.round(keras.backend.clip(y_pred, 0, 1)))
        precision = true_positives / (predicted_positives + keras.backend.epsilon())
        return precision

    precision = precision(y_true, y_pred)
    recall = recall(y_true, y_pred)
    f1 = 2*((precision*recall)/(precision+recall+keras.backend.epsilon()))
    return f1

def f1_score_multiclass(y_true, y_pred):
    f1 = f1_score(y_true, y_pred, average='weighted')
    return f1
```

```python
class MetricsCallback(keras.callbacks.Callback):
    def __init__(self, x_train, y_train, x_val, y_val):
        super().__init__()
        self.x_train = x_train
        self.y_train = y_train
        self.x_val = x_val
        self.y_val = y_val
        self.train_f1_scores = []
        self.val_f1_scores = []

    def after_epoch(self, epoch, logs=None):
        y_train_pred = self.model.predict(self.x_train)
        y_train_pred = np.argmax(y_train_pred, axis=1)
        train_f1 = f1_score_multiclass(self.y_train, y_train_pred)
        self.train_f1_scores.append(train_f1)

        y_val_pred = self.model.predict(self.x_val)
        y_val_pred = np.argmax(y_val_pred, axis=1)
        val_f1 = f1_score_multiclass(self.y_val, y_val_pred)
        self.val_f1_scores.append(val_f1)
```
